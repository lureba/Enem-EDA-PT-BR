## 3. Limpeza e tratamento de dados
    Na etapa de tratamento de dados, realizaremos as seguintes ações:

- Eliminação de colunas irrelevantes: Identificaremos e retiraremos colunas que não são pertinentes, focando apenas nas informações cruciais para nossa análise.

### 3.1 Remoção de Colunas
Dentre algumas colunas que iremos resolver, podemos mencionar algumas:
 - **NU_INSCRICAO**:  Esta coluna contém o número de ID da inscrição do aluno. Como essa informação não contribui para a nossa análise e apenas consome memória, será removida.
 - **NU_ANO**:O ano da prova já é conhecido anteriormente, tornando esta coluna redundante para nossa análise.
 - **CO_MUNICIPIO_PROVA**: Em vez de analisar os dados a nível municipal, é mais relevante e prático focar nas unidades federativas (estados), o que simplifica nossa análise.

### 3.3 Atribuindo os valores

```python

##Categorizando as idades
Answers_Age = {(1,2,3,4):('<=19, Adolescente'),
               (5,6,7,8,9):('20 - 25 , Jovens Adultos'),
               (10,11,12,13,14):('26 - 44, Adulto'),
               (15,16,17,18):('45 - 65, Meia-idade'),
               (19,20):('<66, Idosos')}
##Categorizando o Estado Cívil
Answer_CivilState= {0:'Não Informado',
                    1:'Solteiro(a)',
                    2:'Casado(a)',
                    3:'Divorciado(a)',
                    4:'Viúvo(a)'}
##Categorizando a raça
Answer_Race = {0:'Não declarado',
               1:'Branca',
               2:'Preta',
               3:'Parda',
               4:'Amarela',
               5:'Indigena',
               6:'Não dispõe da informação'}
##Categorizando tipo de escola
Answer_School = {1:'Não Respondeu',
                 2:'Pública',
                 3:'Privada'}
##Categorizando escolaridade pai,mãe
Answer_Q012 = {"A":'Nunca Estudou',
               "B":'Educação Infantil',
               "C":'Fundamental 1',
               "D":'Fundamental 2',
               "E":'Ensino Médio',
               "F":'Graduação Universitária',
               "G":'Pós graduação',
               "H":'Não sei'}
##Categorizando numero de celulares,computadores
Answer_CellComp = {"A":"Não",
                   "B":"Sim, um",
                   "C":"Sim, dois",
                   "D":"Sim, três",
                   "E":"Sim, quatro ou mais"}
##Categorizando o  acesso a internet
Answer_Internet = {"A":"Não",
                   "B":"Sim"}
## Identificação de Treineiro
Answer_T = {0:"Não",
            1:"Sim"}
##Categorizando a renda famíliar
Answer_Income = {'A':'Nenhuma Renda',
                 'B':'Ate 1.212,00',
                 'C':'De 1.212,01 ate 1.818,00',
                 'D':'De 1.818,01 ate 2.424,00',
                 'E':'De 2.424,01 ate 3.030,00',
                 'F':'De 3.030,01 ate 3.636,00',
                 'G':'De 3.636,01 ate 4.848,00',
                 'H':'De 4.848,01 ate 6.060,00',
                 'I':'De 6.060,01 ate 7.272,00',
                 'J':'De 7.272,01 ate 8.484,00',
                 'K':'De 8.484,01 ate 9.696,00',
                 'L':'De 9.696,01 ate 10.908,00',
                 'M':'De 10.908,01 ate 12.120,00',
                 'N':'De 12.120,01 ate 14.544,00',
                 'O':'De 14.544,01 ate 18.180,00',
                 'P':'De 18.180,01 ate 24.240,00',
                 'Q':'Acima de 24.240,00'}
```

```python
t['Nota_geral'] = (t['Nota_cn'] + t['Nota_ch'] + t['Nota_lc'] + t['Nota_mt'] + t['Nota_redacao'])/5
## Criação do valor da nota geral, que será a média das outras notas
```

```python
## Atribuindo os valores a nossas colunas
t['Faixa_etaria'] = t['Faixa_etaria'].replace(Answers_Age)
t['Estado_civil'] = t['Estado_civil'].replace(Answer_CivilState)
t['Cor_raca'] = t['Cor_raca'].replace(Answer_Race)
t['Escola'] = t['Escola'].replace(Answer_School)
t['Escolaridade_pai'] = t['Escolaridade_pai'].replace(Answer_Q012)
t['Escolaridade_mae'] = t['Escolaridade_mae'].replace(Answer_Q012)
t['Celular'] = t['Celular'].replace(Answer_CellComp)
t['Computador'] = t['Computador'].replace(Answer_CellComp)
t['Internet'] = t['Internet'].replace(Answer_Internet)
t['Treineiro'] = t['Treineiro'].replace(Answer_T)
t['Renda_familiar'] = t['Renda_familiar'].replace(Answer_Income)
```

### 3.4 Manipulação para memória

```python
t[conv2] = t[conv2].astype('category')
```
